```python
import pandas as pd

# Load dataset
df = pd.read_csv("gold-dataset-cleaned.csv")

# Show basic statistics and info
print("📊 Dataset Info:")
print(df.info())
print("\n🔎 Missing values per column:")
print(df.isnull().sum())
print("\n📈 Basic statistics:")
print(df.describe(include="all"))

# Ensure new_date is parsed as datetime
df["new_date"] = pd.to_datetime(df["new_date"], errors="coerce")

# Get first and last valid dates
first_date = df["new_date"].min()
last_date = df["new_date"].max()

print("\n✅ Date range in new_date column:")
print("First date:", first_date)
print("Last date:", last_date)

df_sorted = df.sort_values(by="new_date",ascending=True)
display(df_sorted)
```
